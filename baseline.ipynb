```python
import pandas as pd

train = pd.read_csv("data/dataset/train.csv")
test = pd.read_csv("data/dataset/test.csv")

print(train.shape, test.shape)
```

```python
X = train.drop("demand", axis=1)
y = train["demand"]
```

```python
for df in [X, test]:
    df["timestamp"] = pd.to_datetime(df["timestamp"], format="%H:%M")

    df["hour"] = df["timestamp"].dt.hour
    df["minute"] = df["timestamp"].dt.minute

    df.drop("timestamp", axis=1, inplace=True)
```

```python
combined = pd.concat([X, test], axis=0)
```

```python
from sklearn.preprocessing import LabelEncoder

for col in combined.columns:
    if combined[col].dtype == "object":
        le = LabelEncoder()
        combined[col] = le.fit_transform(combined[col].astype(str))
```

```python
X = combined[:len(train)]
test_processed = combined[len(train):]
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

model.fit(X, y)
```

```python
preds = model.predict(test_processed)
```

```python
submission = pd.DataFrame({
    "Index": test["Index"],
    "Demand": preds
})

submission.to_csv("submission_v1.csv", index=False)
```
